```python
import numpy as np
import random
import pandas as pd
```

```python
x=pd.read_csv('Diabaties.csv')
x
```

```python
del(x['Unnamed: 0'])
```

```python
x.head(2)
```

```python
samples=x.iloc[:,0].values
labels=x.iloc[:,1].values
```

# Keras accepts arrays

```python
array_samples=np.array(samples)
array_labels=np.array(labels)
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
scalar=MinMaxScaler(feature_range=(0,1))
```

```python
samples_scalar=scalar.fit_transform((array_samples).reshape(-1,1))
```

```python
print(samples_scalar) #
```

```python
trainx,testx,trainy,testy=train_test_split(samples_scalar,array_labels,test_size=0.2,random_state=0)
trainx,trainy
```

# Importing Keras to train the model

```python
from keras import Sequential
from keras import models,layers,optimizers
```

```python
network=Sequential()
```

```python
network.add(layers.Dense(160,activation='relu',input_shape=(1,)))
network.add(layers.Dense(32,activation='relu'))
network.add(layers.Dense(2,activation='softmax'))
```

```python
network.summary()
```

```python
opt=optimizers.adam(lr=0.0001)
network.compile(optimizer=opt,loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

- By setting the verbose to 0 1 or 2, it denotes how you want to see the progess of each  batch
- verbose 0 shows you nothing
- verbose 1 shows you animated lines like [========================]
- verbose 2 shows you number of epochs like 1/20 Epochs

```python
network.fit(trainx,trainy,batch_size=10,epochs=20,verbose=1)
```
