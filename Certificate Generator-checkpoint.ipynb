```python
# imports
import cv2
# Import QRCode from pyqrcode

import qrcode
from PIL import Image, ImageDraw, ImageFont
  
def certificate_generator(names: list, certificate: str, font_path: str,path="E:/projects/dose/certificated"):
   
    for name in names:
          
        # adjust the position according to 
        # your sample
        text_y_position = 900 
   
        # opens the image
        img = Image.open(certificate, mode ='r')
          
        # gets the image width
        image_width = img.width
          
        # gets the image height
        image_height = img.height 
   
        # creates a drawing canvas overlay 
        # on top of the image
        draw = ImageDraw.Draw(img)
   
        # gets the font object from the 
        # font file (TTF)
        font = ImageFont.truetype(
            font_path,
            200 # change this according to your needs
        )
   
        # fetches the text width for 
        # calculations later on
        text_width, _ = draw.textsize(name, font = font)
   
        draw.text(
            (
                # this calculation is done 
                # to centre the image
                (image_width - text_width) / 2,
                400
            ),
            name,
            font = font,fill="#000000"  )
   
        # saves the image in png format
        img.save(f"{path}/{name}.png")
        
       
        img2 = qrcode.make(name)
        qr = qrcode.QRCode(
            version=1,
            error_correction=qrcode.constants.ERROR_CORRECT_H,
            box_size=10,
            border=4,
        )
        img2 = qr.make_image(fill_color="black", back_color="#edc09a").convert('RGB')
        img2.save(f"{path}/a.png")
        s_img = cv2.imread(f"{path}/a.png")
        s_img = cv2.resize(s_img,(300,300))
        l_img = cv2.imread(f"{path}/{name}.png")
        x_offset=y_offset=50
        l_img[y_offset+780:y_offset+780+s_img.shape[0], x_offset+750:x_offset+750+s_img.shape[1]] = s_img
        cv2.imwrite(f"{path}/{name}.png",l_img)

        
# Driver Code

   
# some example of names
NAMES = ["Ahmed Raza Khanzada","Syed Muhammad Hassan"]

# path to font
FONT = "C:/WINDOWS/FONTS/ITCEDSCR.ttf"

# path to sample certificate
CERTIFICATE = "E:/projects/dose/certificated/Certificate.png"
path = "E:/projects/dose/certificated"
certificate_generator(NAMES, CERTIFICATE, FONT,path=path)
```
