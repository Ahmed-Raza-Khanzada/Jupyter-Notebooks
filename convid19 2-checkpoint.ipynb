```python
def predict(state):
    confirmed = state.groupby("date").sum()["total"].reset_index()
    confirmed.columns = ["ds","y"]
    confirmed["ds"] = pd.to_datetime(confirmed["ds"])
    model1 =Prophet()
    model1.fit(confirmed)
    future_cases=model1.make_future_dataframe(periods=20)
    forecast_cases=model1.predict(future_cases)
    cases = forecast_cases.loc[:,['ds','trend']]
    cases = cases[cases['trend']>0]
    cases.columns = ['Date','cases']
    deaths = state.groupby("date").sum()["Expired"].reset_index()
    deaths = deaths.replace(0,1)
    deaths.columns = ["ds","y"]
    deaths["ds"] = pd.to_datetime(deaths["ds"])
    model2 = Prophet()
    model2.fit(deaths)
    future_deaths=model2.make_future_dataframe(periods=20)
    forecast_deaths=model2.predict(future_deaths)
    death = forecast_deaths.loc[:,['ds','trend']]
    death = death[death['trend']>0]
    death.columns = ['Date','deaths']
    recover = state.groupby("date").sum()["discharged"].reset_index()
    recover.columns = ["ds","y"]
    recover["ds"] = pd.to_datetime(recover["ds"])
    model3 = Prophet()
    model3.fit(recover)
    future_recover=model3.make_future_dataframe(periods=20)
    forecast_recover=model3.predict(future_recover)
    recover = forecast_recover.loc[:,['ds','trend']]
    recover = recover[recover['trend']>0]
    recover.columns = ['Date','recover']
    prediction = cases
    prediction["deaths"] = death[["deaths"]]
    prediction["recovered"] = recover[["recover"]]
    prediction["active"] = cases["cases"]-recover["recover"]
    prediction = prediction.replace(np.nan,int(0))
    return prediction
    
```

```python
sindh_pred = predict(sindh)
sindh_pred["state"] = "sindh"
punjab_pred = predict(punjab)
punjab_pred["state"] = "punjab"
balochistan_pred = predict(balochistan)
balochistan["state"] = "balochistan"
kp_pred = predict(kp)
kp_pred["state"] = "kpk"
gb_pred = predict(gb)
gb_pred["state"] = "gb"
ict_pred = predict(ict)
ict_pred["state"] = "ict"
ajk_pred = predict(ajk)
ajk_pred["state"] = "ajk"
```
