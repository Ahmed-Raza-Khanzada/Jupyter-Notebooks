```python
from prettytable import PrettyTable
class Cart:#cart
    def __init__(self):
        self.cart1=[]
        self.cart2= []
        self.qty1 = []
        self.price=0
        self.option1 = []
    def sub_category(self):
        self.sub = {"1":{"1.1":["car",100],"1.2":["plane",250],"1.3":["truck",300]}
                    ,"2":{"2.1":["Table",500],"2.2":["chair",300],"2.3":["sofa",1000]}
                    ,"3":{"3.1":["tv",1000],"3.2":["Mobile",2000],"3.3":["oven",500]}}
    
        return self.sub
class Store(Cart):
    def option(self):
        while True:
            try:
                print("===========================================================================================================")
                print("***********************************************************************************************************")
                self.option2 = input("Press 1 for Purchasing Toys\nPress 2 for Purchasing Furniture\nPress 3 Purchasing Gadgets\nPress 4 for exit from their")
                if self.option2=="4":
                    self.expense()
                    break
                print("***********************************************************************************************************")
                print("===========================================================================================================")
                self.option1.append(self.option2)
                self.display()
            except ValueError:
                print("Sorry, I didn't understand that. pls input given no")
                continue
    def expense(self):
        p = PrettyTable()
        self.total = sum(self.cart2)
        for i,v in zip(self.cart1,self.cart2):
            p.add_row([i,v])
        p.add_row(["___","___"])
        p.add_row(["Total",self.total])
        p.add_row(["Discount","7%"])
        p.add_row(["___","___"])
        p.add_row(["Your bill is ",((self.total)-(self.total)*0.07)])
        print(f"{p}\nThank You for shopping\nPlease come again")
    def display(self):
        self.c = ("1","2","3","4","5")
        while True:
            if self.option1[-1] in self.c:
                if self.option1[-1]=="1":
                    print("===========================================================================================================")
                    print("***********************************************************************************************************")
                    self.d = self.sub_category().get(str(self.option1[-1]))
                    self.l = len(self.d)
                    self.j = 0
                    print("we have the following items in our store")
                    for i in self.d:
                        print(f"press {i} for {self.d.get(i)}")
                        self.j += 1
                    print("press 4 for exit\npress 5 for main page")
                    self.select= str(input("Selct the product"))
                    if (self.select=="1.1")or(self.select=="1.2")or(self.select=="1.3") :
                        self.m = self.d.get(self.select)
                        self.cart1.append(self.m[0])
                        self.qty = int(input("enter qty"))
                        self.cart2.append(self.m[1]*self.qty)
                        self.qty1.append(self.qty)
                        break
                    elif self.select=="4":
                        break
                    elif self.select=="5":
                        self.option()
                        break
                    else:
                        print("wrong input")
                    print("***********************************************************************************************************")
                    print("===========================================================================================================")
                elif self.option1[-1]=="2":
                    print("===========================================================================================================")
                    print("***********************************************************************************************************")
                    self.d = self.sub_category().get(str(self.option1[-1]))
                    self.l = len(self.d)
                    self.j = 0
                    print("we have the following items in our store")
                    for i in self.d:
                        print(f"press {i} for {self.d.get(i)}")
                        self.j += 1
                    print("press 4 for exit\npress 5 for main page")
                    self.select= str(input("Selct the product"))
                    if (self.select=="2.1")or(self.select=="2.2")or(self.select=="2.3") :
                        self.m = self.d.get(self.select)
                        self.cart1.append(self.m[0])
                        self.qty = int(input("enter qty"))
                        self.cart2.append(self.m[1]*self.qty)
                        self.qty1.append(self.qty)
                        break
                    elif self.select=="4":
                        break
                    elif self.select=="5":
                        self.option()
                        break
                    else:
                        print("wrong input")
                    print("***********************************************************************************************************")
                    print("===========================================================================================================")
                elif self.option1[-1]=="3":
                    print("===========================================================================================================")
                    print("***********************************************************************************************************")
                    self.d = self.sub_category().get(str(self.option1[-1]))
                    self.l = len(self.d)
                    self.j = 0
                    print("we have the following items in our store")
                    for i in self.d:
                        print(f"press {i} for {self.d.get(i)}")
                        self.j += 1
                    print("press 4 for exit\npress 5 for main page")
                    self.select= str(input("Selct the product"))
                    if (self.select=="3.1")or(self.select=="3.2")or(self.select=="3.3") :
                        self.m = self.d.get(self.select)
                        self.cart1.append(self.m[0])
                        self.qty = int(input("enter qty"))
                        self.cart2.append(self.m[1]*self.qty)
                        self.qty1.append(self.qty)
                        break
                    elif self.select=="4":
                        break
                    elif self.select=="5":
                        self.option()
                        break
                    else:
                        print("wrong input")
                    print("***********************************************************************************************************")
                    print("===========================================================================================================")
                elif self.option1[-1]=="4":
                    break
                elif self.option1[-1]=="5":
                    self.option()
                    break
                else:
                    continue
            else:
                continue
        print("IN YOUR CART")
        for i,v in zip(self.cart1,self.cart2):
            print(f"{i}----->{v} RS")
    
```
