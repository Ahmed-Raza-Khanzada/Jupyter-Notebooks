```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import os
import glob
import cv2
from tensorflow.keras.layers import *
from tensorflow.keras.models import *
from tensorflow.keras.applications import *
from tensorflow.keras.optimizers import *
from tensorflow.keras.regularizers import *

```

```python
def load_data(path):
    train_x = []
    test_x = []
    print("Loading Dataset")
    for file in os.listdir(path):
        if file=="train":
            path1 = os.path.join(path,file)
            img_path = os.path.join(path1.replace("\\","/"),'*g')
            files = glob.glob(img_path)
            for f1 in files:
                img = cv2.imread(f1,cv2.COLOR_BGR2RGB)
                img = cv2.resize(img,(224,224))
                train_x.append(img)
        elif file=="test":
            path1 = os.path.join(path,file)
            img_path = os.path.join(path1.replace("\\","/"),'*g')
            files = glob.glob(img_path)
            for f1 in files:
                img = cv2.imread(f1,cv2.COLOR_BGR2RGB)
                img = cv2.resize(img,(224,224))
                test_x.append(img)
        elif file=="labels.csv":
                labels = pd.read_csv(path+"/"+file)
        else:
            pass
    print("Completed")
    return train_x,test_x,labels
```

```python
x_train,x_test,labels = load_data(path="E:/Dog Breed Dataset")
```

```python
def get_features(MODEL, data):
    cnn_model = MODEL(include_top=False, input_shape=(width, width, 3), weights='imagenet')
    
    inputs = Input((width, width, 3))
    x = inputs
    x = Lambda(preprocess_input, name='preprocessing')(x)
    x = cnn_model(x)
    x = GlobalAveragePooling2D()(x)
    cnn_model = Model(inputs, x)
    features = cnn_model.predict(data, batch_size=64, verbose=1)
    return features

```
