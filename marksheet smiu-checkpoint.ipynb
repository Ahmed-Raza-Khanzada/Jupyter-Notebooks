```python
import os
def read_data(x):
    f = open(x)
    file_data = f.read()
    return file_data
def notepad():
    path ="E:/notepad/"
    while True:
        user = input("if you want to create snd write something in file Press 1\nTo append something in file Press 2\nTo delete something in file press 3")
        if user!=[]:
            if user in ("1","2","3"):
                break
            else:
                continue
        else:
            print("please enter something")
            continue
    if user=="1":
        while True:
            string = input("Enter string u want to enter")
            if string!="":
                break
            else:
                print("please enter something")
                continue
        if os.path.exists(path):
            with open(f"{path}/file.txt","w") as f:
                f.write(string)
            file = open(f"{path}/file.txt",'r')
            print(read_data(f"{path}/file.txt"))
            file.close()
        else:
            os.makedirs(path)
            with open(f"{path}/file.txt","w") as f:
                f.write(string)
            file = open(f"{path}/file.txt",'r')
            print(read_data(f"{path}/file.txt"))
            file.close()
    elif user=="2":
        while True:
            string = input("Enter string u want to enter")
            if string!="":
                break
            else:
                print("please enter something")
                continue
        if os.path.exists(path):
            with open(f"{path}/file.txt","a") as f:
                f.write("\n"+string)
            file = open(f"{path}/file.txt",'r')
            print(read_data(f"{path}/file.txt"))
            file.close()
        else:
            os.makedirs(path)
            with open(f"{path}/file.txt","a+") as f:
                f.write("\n"+string)
            file = open(f"{path}/file.txt",'r')
            print(read_data(f"{path}/file.txt"))
            file.close()
    elif user=="3":
        while True:
            string = input("Enter string u want to enter")
            if string!="":
                break
            else:
                print("please enter something")
                continue
        if os.path.exists(path):
            x = read_data(f"{path}/file.txt")
            f = open(f"{path}/file.txt",'w')
            y = x.replace(string,'')
            f.write(y)
            f.close()
            print(read_data(f"{path}/file.txt"))

        else:
            print("File is not exist")
        
```
