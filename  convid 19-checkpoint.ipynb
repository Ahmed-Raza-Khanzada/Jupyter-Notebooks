```python
import pandas as pd  
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as seabornInstance 
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn import metrics
import seaborn as sns
```

```python
data = pd.read_excel("E:/cat and dog/COVID_FINAL_DATA.xlsx",parse=["Date"])
```

```python
data = data.rename(columns={"Unnamed: 0":"unname","Cumulative":"total","Cumulative  Test positive":"new",
                     "Cumulative  tests performed":"test perform","Date":"date",
                    "Discharged":"discharged","expire":"expired",
                     "Home Quarantine":"home qurantine","New  (last 24 hrs)":"24 hour",
                     "Region":"region","Still admitted":"still admitted","Tests  performed in last 24 hours":"test in 24 hour"})
```

```python
cols = data.columns
cols[11]
```

```python
data.head(10).style.background_gradient(cmap='Greens')
```

```python
data = data.replace(np.nan,0)
```

```python
dates = data["date"]
```

```python
data
```

```python
data = data.drop(data.columns[data.columns.str.contains("unname",case=False)],axis=1)
data = data.drop(data.columns[data.columns.str.contains("test perform",case=False)],axis=1)
data = data.drop(data.columns[data.columns.str.contains("home qurantine",case=False)],axis=1)
data = data.drop(data.columns[data.columns.str.contains("test in 24 hour",case=False)],axis=1)
data = data.drop(data.columns[data.columns.str.contains("total",case=False)],axis=1)
data = data.drop(data.columns[data.columns.str.contains("24 hour",case=False)],axis=1)
data = data.drop(data.columns[data.columns.str.contains("still admitted",case=False)],axis=1)

```

```python
data['date'] = pd.to_datetime(data['date']).dt.date
data
```

```python
data = data.rename(columns={"new":"total"})
```

```python
data.tail(10)
```

```python
data = data.sort_values(["region","date"])
```

```python
data
```

```python
plt.figure(figsize = (10,15))
plt.xticks(rotation=90,fontsize=12)
plt.yticks(fontsize=15)
plt.xlabel("date",fontsize=30)
plt.ylabel("total cases",fontsize=30)
plt.title("Pakistan's States Cases Over Time",fontsize=35,color="r")
total_cases = data.groupby("date")["date","total"].sum().reset_index()
total_cases["date"]= pd.to_datetime(total_cases["date"])
ax = sns.pointplot(x=total_cases.date.dt.date,y=total_cases.total,color="g")
ax.set(xlabel="Dates",ylabel="Total Cases")
```

```python
head1 = data[data["date"]==data["date"].max()]
state = head1.groupby("region")["total","discharged","Expired"].sum().reset_index()
```

```python
head1
```

```python
top_cases = head1.groupby("region")["total"].sum().sort_values(ascending=False).reset_index()
```

```python
plt.figure(figsize = (15,10))
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel("Total_cases",fontsize=30)
plt.ylabel("States",fontsize=30)
plt.title("Pakistan's States Total Cases",fontsize=35,color="g")
ax = sns.barplot(x=top_cases.total,y=top_cases.region)
for i, (value,name) in enumerate(zip(top_cases.total,top_cases.region)):
    ax.text(value,i-0.5,f"{value:,.0f}",size=10,ha="left",va="center")
ax.set(xlabel="Total Cases",ylabel="States")
```

```python
top_deaths = head1.groupby("region")["Expired"].sum().sort_values(ascending=False).reset_index()
```

```python
plt.figure(figsize = (15,10))
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel("Total_cases",fontsize=30)
plt.ylabel("States",fontsize=30)
plt.title("Pakistan's States Total Deaths",fontsize=35,color="g")
ax = sns.barplot(x=top_deaths.Expired,y=top_deaths.region)
for i, (value,name) in enumerate(zip(top_deaths.Expired,top_deaths.region)):
    ax.text(value,i-0.5,f"{value:,.0f}",size=10,ha="left",va="center")
ax.set(xlabel="Total Deaths",ylabel="States")
```

```python
top_discharged = head1.groupby("region")["discharged"].sum().sort_values(ascending=False).reset_index()
```

```python
plt.figure(figsize = (15,10))
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel("Total_cases",fontsize=30)
plt.ylabel("States",fontsize=30)
plt.title("Pakistan's States Total Recovered",fontsize=35,color="g")
ax = sns.barplot(x=top_discharged.discharged,y=top_discharged.region)
for i, (value,name) in enumerate(zip(top_discharged.discharged,top_discharged.region)):
    ax.text(value,i-0.5,f"{value:,.0f}",size=10,ha="left",va="center")
ax.set(xlabel="Total Recovered",ylabel="States")
```

```python
sindh = data[data.region=="Sindh"]
sindh = sindh.groupby("date")["total","discharged","Expired"].sum().reset_index()
punjab = data[data.region=="Punjab"]
punjab = punjab.groupby("date")["total","discharged","Expired"].sum().reset_index()
balochistan = data[data.region=="Balochistan"]
balochistan = balochistan.groupby("date")["total","discharged","Expired"].sum().reset_index()
gb = data[data.region=="GB"]
gb = gb.groupby("date")["total","discharged","Expired"].sum().reset_index()
kp = data[data.region=="KP"]
kp = kp.groupby("date")["total","discharged","Expired"].sum().reset_index()
ict = data[data.region=="ICT"]
ict = ict.groupby("date")["total","discharged","Expired"].sum().reset_index()
ajk = data[data.region=="AJK"]
ajk = ajk.groupby("date")["total","discharged","Expired"].sum().reset_index()
```

```python
fig,ax= plt.subplots()
ax.plot(sindh.index,sindh.total,color="green", label="Sindh",marker="*")
ax.plot(punjab.index,punjab.total,color="blue", label="Punjab",marker="*")
ax.plot(kp.index,kp.total,color="orange", label="Kp",marker="*")
ax.plot(balochistan.index,balochistan.total,color="brown", label="Balochistan",marker="*")
ax.plot(gb.index,gb.total,color="pink", label="GB",marker="*")
ax.plot(ict.index,ict.total,color="red", label="ict",marker="*")
ax.plot(ajk.index,ajk.total,color="purple", label="ajk",marker="*")
ax.set_ylabel('No of Cases')
ax.set_xlabel('No of days')
ax.set_title('Pakistan states by cases')
plt.show()
```
