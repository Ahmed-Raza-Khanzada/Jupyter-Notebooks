```python
import matplotlib.pyplot as plt
import csv
x = []
y = []

with open("Downloads/test.txt","r") as csvfile:
    plots = csv.reader(csvfile)
    for col in plots:
        x.append(col[0])
        y.append(col[1])
plt.plot(x,y,label= "File",marker = ".")
plt.xlabel("X-axis")
plt.ylabel("y-axis")
plt.title("text file")
plt.legend()
plt.show()
```
