```python
import matplotlib.pyplot as plt
from skimage.feature import hog
from skimage import data, color, exposure
from skimage.feature import local_binary_pattern
from skimage.color import label2rgb
import pywt
import pywt.data
METHOD = 'uniform'
# settings for LBP
radius = 3
n_points = 8 * radius

import os
import array
import pandas as pd
os.chdir('C:/Users/Mr. Irfan/Documents/Face_Recoginton_HOG_DWT_LBT/')
import time
from PIL import Image
i=0
columnNames = list()

#myFileList1 = createFileList('C:/Users/Mr. Irfan/Documents/Sindhi_ML_project/cake_slices')
columnNames.append('label')
for i in range(4356):
    pixel = 'pixel'
    pixel += str(i)
    columnNames.append(pixel)
train_data = pd.DataFrame(columns = columnNames)
start_time = time.time()
c=0
for foldr,img in zip(myFileListlabel,myFileList):
    label=foldr.split("//")
    i=i+1
    t = i
    from scipy.misc import imread
    image = imread(img)
    image = color.rgb2gray(image)
    #image = color.rgb2gray(data.astronaut())


    fd, hog_image = hog(image, orientations=8, pixels_per_cell=(3, 3),
                    cells_per_block=(1, 1), visualise=True)
     
    
    # Rescale histogram for better display
    hog_image_rescaled = exposure.rescale_intensity(hog_image, in_range=(0, 0.02))
    
    # local binary pattarn
    lbp = local_binary_pattern(hog_image_rescaled, n_points, radius, METHOD)
   
    # DWT
    coeffs2 = pywt.dwt2(lbp, 'bior1.3')
    LL, (LH, HL, HH) = coeffs2
  
    
    data = []
    data.append(label[1])
    for y in range(66):
        for x in range(66):
            data.append(HL[y,x])
  
        
    c=c+1
    train_data.loc[i] = [data[k] for k in range(4357)]
    
train_data.to_csv("hog_dwt_lbp.txt",index = False,encoding='utf-8')
print ("Hog ,discreate Wavelet transform and local binary pattarn Done1")
    
    
```

```python
from skimage.transform import rotate
from skimage.feature import local_binary_pattern
from skimage import data
from skimage.color import label2rgb
METHOD = 'uniform'
# settings for LBP
radius = 3
n_points = 8 * radius


def overlay_labels(image, lbp, labels):
    mask = np.logical_or.reduce([lbp == each for each in labels])
    return label2rgb(mask, image=image, bg_label=0, alpha=0.5)


def highlight_bars(bars, indexes):
    for i in indexes:
        bars[i].set_facecolor('r')


#image = data.load('brick.png')
#image = color.rgb2gray(data.astronaut())
lbp = local_binary_pattern(image, n_points, radius, METHOD)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 4), sharex=True, sharey=True)

ax1.axis('off')
ax1.imshow(image, cmap=plt.cm.gray)
ax1.set_title('Input image')


ax2.axis('off')
ax2.imshow(lbp, cmap=plt.cm.gray)
ax2.set_title('Histogram of Oriented Gradients')
plt.show()


def hist(ax, lbp):
    n_bins = int(lbp.max() + 1)
    return ax.hist(lbp.ravel(), normed=True, bins=n_bins, range=(0, n_bins),
                   facecolor='0.5')


# plot histograms of LBP of textures
fig, (ax_img, ax_hist) = plt.subplots(nrows=2, ncols=3, figsize=(9, 6))
plt.gray()

titles = ('edge', 'flat', 'corner')
w = width = radius - 1
edge_labels = range(n_points // 2 - w, n_points // 2 + w + 1)
flat_labels = list(range(0, w + 1)) + list(range(n_points - w, n_points + 2))
i_14 = n_points // 4            # 1/4th of the histogram
i_34 = 3 * (n_points // 4)      # 3/4th of the histogram
corner_labels = (list(range(i_14 - w, i_14 + w + 1)) +
                 list(range(i_34 - w, i_34 + w + 1)))

label_sets = (edge_labels, flat_labels, corner_labels)

for ax, labels in zip(ax_img, label_sets):
    ax.imshow(overlay_labels(image, lbp, labels))

for ax, labels, name in zip(ax_hist, label_sets, titles):
    counts, _, bars = hist(ax, lbp)
    highlight_bars(bars, labels)
    ax.set_ylim(top=np.max(counts[:-1]))
    ax.set_xlim(right=n_points + 2)
    ax.set_title(name)

ax_hist[0].set_ylabel('Percentage')
for ax in ax_img:
    ax.axis('off')
```
