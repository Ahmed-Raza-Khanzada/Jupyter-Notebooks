# Concatenation 

```python
ar1 = np.arange(20).reshape(5,4)
ar2 = np.arange(21,41).reshape(5,4)
print(ar1)
print("**************")
print(ar2)
```

```python
np.concatenate((ar1,ar2),axis=0)
```

```python
np.concatenate((ar1,ar2),axis=1)
```
