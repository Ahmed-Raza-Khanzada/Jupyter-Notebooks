# TESTING MODEL ACCURACY ON ANN

```python
import numpy as np
import pandas as pd
import tensorflow
import os
from skimage import io,color
import cv2
import glob
from tensorflow.keras import layers
from tensorflow.keras.layers import Dense,Input
from tensorflow.keras.layers import Conv2D,MaxPooling2D,Dropout,Flatten
from tensorflow.keras.models import Sequential
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
y = pd.read_csv("E:/urdu_letter/labels.csv",header = None)
```

```python

img_dir = "E:/urdu_letter/updated_urdu_letter" # Enter Directory of all images 
data_path = os.path.join(img_dir,'*g')
files = glob.glob(data_path)
x = []
for f1 in files:
    img = cv2.imread(f1,cv2.IMREAD_GRAYSCALE)
    x.append(img)
```

```python
import matplotlib.pyplot as plt
x = np.array(x)
plt.imshow(x[0])
plt.show()
```

```python
y = np.array(y)
y = y.reshape(24116,)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,shuffle=True,test_size = 0.3,random_state=42)
```

```python
plt.imshow(x_train[1002])
```

```python
y_train[1002]
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
x_train = x_train.reshape(16881,1024)
x_train = x_train.astype("float")/255
x_test = x_test.reshape(7235,1024)
x_test = x_test.astype("float")/255
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.fit_transform(y_test)
y_train.shape
```

```python
from tensorflow.keras.utils import to_categorical
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
y_train.shape
```

```python
x_val=x_test[0:2235]
y_val = y_test[0:2235]
x_test = x_test[2235:]
y_test = y_test[2235:]
```

```python
#from tensorflow.keras.optimizers import SGD
#sgd = SGD(lr = 0.01,momentum = 0.9,nesterov = True)
model = Sequential()
model.add(layers.Dense(1020,activation= "relu",input_shape= (1024,)))
model.add(layers.Dense(900,activation = "relu"))
model.add(layers.Dense(800,activation="relu"))
model.add(Dropout(0.5))
#model.add(layers.Dense(700,activation="relu"))
model.add(layers.Dense(38,activation="softmax"))
model.compile(optimizer = "adam" ,loss = "categorical_crossentropy" , metrics = ["accuracy"],lr = 0.01,momentum = 0.7)
```

```python
model.fit(x_train,y_train,batch_size = 500,epochs = 22,validation_data=(x_val,y_val))
```

```python
a=model.predict(x_test)
a[0]
```

```python
loss,acc = model.evaluate(x_test,y_test,verbose = 0)
print("               Using ANN                         ")
print(f"loss on testing is {loss}\naccuracy on testing is {acc}")
```

# TESTING MODEL ACCURACY ON CNN

```python
y = pd.read_csv("E:/urdu_letter/labels.csv",header = None)
```

```python
img_dir = "E:/urdu_letter/updated_urdu_letter" # Enter Directory of all images 
data_path = os.path.join(img_dir,'*g')
files = glob.glob(data_path)
x = []
for f1 in files:
    img = cv2.imread(f1,cv2.IMREAD_GRAYSCALE)
    x.append(img)
```

```python
import matplotlib.pyplot as plt
x = np.array(x)
plt.imshow(x[10000])
plt.show()
```

```python
y = np.array(y)
y = y.reshape(24116,)
x.ndim
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,shuffle=True,test_size = 0.3,random_state=20)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.fit_transform(y_test)
```

```python
from tensorflow.keras.utils import to_categorical
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

```python
#x_train = x_train.reshape(16881,1024)
x_train = x_train.astype("float")/255
#x_test = x_test.reshape(7235,1024)
x_test = x_test.astype("float")/255
x_test.shape
```

```python
from tensorflow.keras.models import Model
x_train = x_train.reshape(16881,32,32,1)
x_test = x_test.reshape(7235,32,32,1)
```

```python
x_val=x_test[0:2235]
y_val = y_test[0:2235]
x_test = x_test[2235:]
y_test = y_test[2235:]
```

```python
inpx = (32,32,1)
inpx = Input(shape=inpx)
layer1 = Conv2D(100,kernel_size=(3,3),activation="relu")(inpx)
layer2 = Conv2D(68,kernel_size = (3,3),activation="relu")(layer1)
layer8 = Conv2D(100,kernel_size = (3,3),activation="relu")(layer2)
layer3 = MaxPooling2D(pool_size= (3,3))(layer8)
#to prevent overfitting
#layer4 = Dropout(0.5)(layer3)
layer5 = Flatten()(layer3)
layer6 = Dense(250,activation="sigmoid")(layer5)
layer7 = Dense(38,activation="softmax")(layer6)
```

```python
model = Model([inpx],layer7)
model.compile(optimizer = "adam",loss = tensorflow.keras.losses.categorical_crossentropy,metrics= ["accuracy"],lr=0.01)
```

```python
x_test.shape
```

```python
model.fit(x_train,y_train,batch_size = 1000,epochs = 15,validation_data=(x_val,y_val))
```

```python
score = model.evaluate(x_test,y_test,verbose = 0)
print(f"loss{score[0]} and accuracy {score[1]}")
```
