```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6]
y = [25,128,40,2,8,0.55]
countries = ["Pakistan","India","America","Saudia","Turkey","Qatar"]
plt.bar(x,y,tick_label=countries,width=0.7,color=["green"])
plt.title("Countries population")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```
