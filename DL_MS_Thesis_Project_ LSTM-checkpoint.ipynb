```

data = data[data['Sentiment'] != 2]
data = data[data['type'] == 3 or]


data['SentimentText'] = data['SentimentText'].apply(lambda x: x.lower())
data['SentimentText'] = data['SentimentText'].apply((lambda x: re.sub('[^a-zA-z0-9\s]','',x)))
#print(data['SentimentText'])
#print(data[ data['Sentiment'] == 1].size)
#print(data[ data['Sentiment'] == 0].size)

for idx,row in data.iterrows():
    row[0] = row[0]

max_fatures = 1000
tokenizer = Tokenizer(num_words=max_fatures, split=' ')
tokenizer.fit_on_texts(data['SentimentText'].values)
#print(tokenizer.word_index)
X = tokenizer.texts_to_sequences(data['SentimentText'].values)
X = pad_sequences(X)

print(X)
```

```

#set the stopwords list
stopwords= set(STOPWORDS)
new_words=('teacher','class','instructor','feedback','course','student','lecture','material','teaching','assignment','classes','assessment','subject','lecturer','students','time','resources','thing','think','teachers','make','take')
new_stopwords=stopwords.union(new_words)
```

```

train_pos = data[ data['Sentiment'] == 1]
train_pos = train_pos['SentimentText']
train_neg = data[ data['Sentiment'] == 0]
train_neg = train_neg['SentimentText']
def wordcloud_draw(data, color = 'black'):
    words = ' '.join(data)
    cleaned_word = " ".join([word for word in words.split()
                            if 'http' not in word
                                and not word.startswith('@')
                                and not word.startswith('#')
                                and word != 'RT'
                            ])
    wordcloud = WordCloud(stopwords=new_stopwords,
                      background_color=color,
                      width=1500,
                      height=1000
                     ).generate(cleaned_word)
    plt.figure(1,figsize=(11, 11))
    plt.imshow(wordcloud)
    plt.axis('off')
    plt.show()
    
print("Positive words")
wordcloud_draw(train_pos,'white')
print("Negative words")
wordcloud_draw(train_neg,'white')
```

```

embed_dim = 64
lstm_out = 196
model = Sequential()
model.add(Embedding(max_fatures, embed_dim,input_length = X.shape[1]))
#model.add(LSTM(lstm_out,dropout=0.2,activation='relu'))
model.add(LSTM(lstm_out, dropout=0.2))
model.add(Dense(2,activation='softmax'))
#model.compile(loss = 'binary_crossentropy', optimizer='adam',metrics = ['accuracy'])
model.compile(loss = 'categorical_crossentropy', optimizer='Adam',metrics = ['accuracy'])
print(model.summary())
```

```
Y = pd.get_dummies(data['Sentiment']).values
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.33)
print(X_train.shape,Y_train.shape)
print(X_test.shape,Y_test.shape)
```

# Hyperparameter Tuning


```

embed_dim = 64
lstm_out = 196
dropout_rate=0.2
optimizer = ['Adagrad', 'Adam', 'Adamax', 'Nadam'] 
colors = ['red', 'green', 'blue', 'black']
markers = ['*', 'o', '^', '+'] 

# Function to create model, required for KerasClassifier
def create_model(activation='softmax', optimizer='Adam', dropout_rate=0.2):
    model = Sequential()
    model.add(Embedding(max_fatures, embed_dim,input_length = X.shape[1]))
    model.add(LSTM(lstm_out, dropout=dropout_rate))
    model.add(Dense(3,activation=activation))
    model.compile(loss = 'categorical_crossentropy', optimizer=optimizer, metrics = ['accuracy'])
    return model
```


**Tuning through the Different optimization algorithms  Adadelta, Adam, Adamax, Nadam**

```
fig, ax = plt.subplots()
for i in range(len(optimizer)):
  model=create_model(activation='softmax', optimizer=optimizer[i], dropout_rate=0.2)  
  batch_size = 64
  x=range(1,31)
  #early_stop = EarlyStopping(monitor='val_loss', patience=20, verbose=0)
  history=model.fit(X_train, Y_train, epochs = 30,validation_data=(X_test, Y_test), batch_size=batch_size, verbose = 0)
  ax.plot(x,history.history['val_acc'], color=colors[i], label=optimizer[i], marker=markers[i])
  ax.set_ylabel('Accuracy')
  ax.set_xlabel('Epochs')
  ax.set_title('Validation Accuracy Vs Epochs (Dropout:0.2 , Activation:softmax)')
  ax.legend(loc='upper left')
```

**Tuning through the Different Activation Function softmax ,softplus , relu, tanh**

```
activation = ['softmax','softplus','sigmoid','relu']  
#activation = ['softmax']  

fig, ax = plt.subplots()
for i in range(len(activation)):
  model=create_model(activation=activation[i], optimizer='Adam', dropout_rate=0.2)  
  batch_size = 64
  x=range(1,31)
  #early_stop = EarlyStopping(monitor='val_loss', patience=20, verbose=1)
  history=model.fit(X_train, Y_train, epochs = 30,validation_data=(X_test, Y_test), batch_size=batch_size, verbose = 0)
  ax.plot(x,history.history['val_acc'], color=colors[i], label=activation[i], marker=markers[i])
  ax.set_ylabel('Accuracy')
  ax.set_xlabel('Epochs')
  ax.set_title('Validation Accuracy Vs Epochs (Dropout:0.2 , optimzer:Adam)')
  ax.legend(loc='upper left')
```

**Tuning through the Different Dropout Rates 0.1, 0.2, 0.3, 0.4**

```
dropout_rate = [0.1, 0.2, 0.3, 0.4]
  
fig, ax = plt.subplots()
for i in range(len(dropout_rate)):
  model=create_model(activation='softmax', optimizer='Adam', dropout_rate=dropout_rate[i])  
  batch_size = 64
  x=range(1,31)
  #early_stop = EarlyStopping(monitor='val_loss', patience=20, verbose=1)
  history=model.fit(X_train, Y_train, epochs = 30,validation_data=(X_test, Y_test), batch_size=batch_size, verbose = 0)
  ax.plot(x,history.history['val_acc'], color=colors[i], label=dropout_rate[i], marker=markers[i])
  ax.set_ylabel('Accuracy')
  ax.set_xlabel('Epochs')
  ax.set_title('Validation Accuracy Vs Epochs(Activation:softmax, optmizer:adam)')
  ax.legend(loc='upper left')
```

```
pos_cnt, neg_cnt, pos_correct, neg_correct = 0, 0, 0, 0
for x in range(len(X_validate)):
    result = model.predict(X_validate[x].reshape(1,X_test.shape[1]),batch_size=1,verbose = 2)[0]
    if np.argmax(result) == np.argmax(Y_validate[x]):
        if np.argmax(Y_validate[x]) == 0:
            neg_correct += 1
        else:
            pos_correct += 1
       
    if np.argmax(Y_validate[x]) == 0:
        neg_cnt += 1
    else:
        pos_cnt += 1


print("postive_acc: %.2f" % (100 * pos_correct/pos_cnt))
print("negative_acc: %.2f" % (100 * neg_correct/neg_cnt))
```
